```python
import xarray as xr
import os
```

```python
inp_dir="C:/Mayank/Projects/Project_Viatris/Data/Rainfall_Temperature/New_IMD/Rainfall/Rainfall1deg"
```

```python
file_ext=['.nc']
```

```python
file_paths=[]
```

```python
for root,dirs,files in os.walk(inp_dir):
    for file_name in files:
        if any(file_name.endswith(ext) for ext in file_ext):
            file_path=os.path.join(root, file_name)
            file_paths.append(file_path)
```

```python
file_paths
```

```python
datasets=[xr.open_dataset(file_path) for file_path in file_paths]
```

```python
merged_ds=xr.concat(datasets,dim='TIME')
```

```python
print(merged_ds)
```

```python
merged_ds.to_netcdf(os.path.join(root, 'rainfall_25deg_1990_.nc'))
```
