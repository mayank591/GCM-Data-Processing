```python
import geopandas as gpd
import pandas as pd
from shapely import wkt
import glob
import os
import csv
import time
```

```python
# inpdir = 'C:/Mayank/Projects/Project_CDRI/Data/DoT/Complete/SiteMaster'
# outdir = 'C:/Mayank/Projects/Project_CDRI/Data/DoT/Complete/SiteMaster_Shape/'
inpdir = 'C:/Mayank/Projects/Project_CDRI/Data/DoT/Complete/TenancyMaster'
outdir = 'C:/Mayank/Projects/Project_CDRI/Data/DoT/Complete/TenancyMaster_Shape/'
```

```python
filelist = glob.glob(os.path.join(inpdir, '*.csv'))
```

```python
filelist
```

```python
for filenames in filelist:
    filen = os.path.basename(filenames).split('.')[0]
#     search_string = "Textbox5"
    search_string = "Textbox30"
    linecount = 0
    with open(filenames, 'r',  encoding='utf-8') as f:
        reader = csv.reader(f)
        for num, row in enumerate(reader):
#             print(num, row)
            if search_string in row:
#                 print(num, row)
                linecount = num-1
    data = pd.read_csv(filenames, skiprows=linecount)
#     print(data.columns)
    data_gdf = gpd.GeoDataFrame(data, geometry = gpd.points_from_xy(data['Longitude'],data['Lattitude']))
    data_gdf.set_crs(epsg=4326, inplace=True)
    data_gdf.to_file(os.path.join(outdir, filen+'.shp'), driver = 'ESRI Shapefile')
```
