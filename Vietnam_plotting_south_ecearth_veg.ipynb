```python
import rioxarray
import xarray as xs
import matplotlib.pyplot as plt
import os
import glob
import datetime
from netCDF4 import Dataset, num2date, date2num
import numpy as np
%matplotlib inline
import pandas as pd
from calendar import isleap
from pandas.tseries.offsets import DateOffset
from pathlib import Path 
import seaborn as sns
```

```python
region='South'
```

```python
folder = "C:/Mayank/Projects/Project_Vietnam/NewData/GCM_Download/ECEARTH_Veg/Merged/Clipped/Unit_Changed/Regional"
```

```python
ds1 = xs.open_dataset(os.path.join(folder, "ssp585_tas_ecearth_veg_final_"+region+".nc"))
```

```python
ds2 = xs.open_dataset(os.path.join(folder, "ssp245_tas_ecearth_veg_final_"+region+".nc"))
```

```python
fig,ax = plt.subplots(figsize=(10,7))
ds1.tas.groupby('time.year').mean(['time','lon','lat']).plot(ax=ax,label='SSP585')
ds2.tas.groupby('time.year').mean(['time','lon','lat']).plot(ax=ax,label='SSP245')
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.0),
          ncol=4, fancybox=True, shadow=True)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,7))
ds1.tas.groupby('time.month').mean(['time','lon','lat']).plot(ax=ax,label='SSP585')
ds2.tas.groupby('time.month').mean(['time','lon','lat']).plot(ax=ax,label='SSP245')
ax.legend(loc='upper right',ncol=4, fancybox=True, shadow=True)
plt.show()
```
